```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('cleaned_data.csv', sep=',')
print('Shape of the data', data.shape)
data.head()
```

## Univariate Analysis

```python
## Histogram && Box Plot 
def plot_histogram_and_boxplot(df, column_name, bins=30):

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Histogram
    sns.histplot(df[column_name].dropna(), bins=bins, kde=True, ax=axes[0])  # Drop NaNs for accurate distribution
    axes[0].set_title(f'Distribution of {column_name}')
    axes[0].set_xlabel(column_name)
    axes[0].set_ylabel('Frequency')
    
    # Box Plot
    sns.boxplot(x=df[column_name].dropna(), ax=axes[1])  # Drop NaNs for accurate box plot
    axes[1].set_title(f'Box Plot of {column_name}')
    axes[1].set_xlabel(column_name)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_histogram_and_boxplot(data, 'Item_MRP')
```

```python
plot_histogram_and_boxplot(data,'Item_Weight')
```

```python
plot_histogram_and_boxplot(data, 'Item_Visibility')
```
